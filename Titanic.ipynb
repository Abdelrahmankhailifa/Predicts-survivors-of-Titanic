```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("E:\\Downloads\\titanic\\train.csv")
test = pd.read_csv("E:\\Downloads\\titanic\\test.csv")
Pass_id = test["PassengerId"]
```

```python
inputs_data = data.drop(data[['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked','Survived']],axis='columns')
target = data.Survived
inputs_data.head()
```

```python
le_sex = LabelEncoder()
inputs_data['n_sex'] = le_sex.fit_transform(inputs_data.Sex)
updated_inputs = inputs_data.drop('Sex',axis='columns')
updated_inputs.Age = inputs_data.Age.fillna(inputs_data.Age.mean())
updated_inputs
```

```python
test1 = test.drop(test[['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked']],axis='columns')
test1.head()
```

```python
le_sex = LabelEncoder()
test1['n_sex'] = le_sex.fit_transform(test1.Sex)
updated_inputs_test = test1.drop('Sex',axis='columns')
updated_inputs_test.Age = test1.Age.fillna(test1.Age.mean())
updated_inputs_test
```

```python
model = tree.DecisionTreeClassifier()
X_train, X_test, y_train, y_test = train_test_split(updated_inputs,target,test_size=0.2)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
sub = model.predict(updated_inputs_test)
sub
```

```python
submission = pd.DataFrame({
        "PassengerId": Pass_id,
        "Survived": sub
    })

submission.to_csv('sub.csv', index=False)
```
